```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
panel_area = 100  # square meters
panel_efficiency = 0.15  # %
latitude = -35.5  # La Pampa, Argentina
days_per_year = 365

# Solar radiation data for La Pampa, Argentina
mean_radiation = 5.0  # kWh/m^2/day
stddev_radiation = 0.5  # kWh/m^2/day

# Run the simulation
results = []
for i in range(1000):
    for x in range(12):
        # Generate a random sample for solar radiation on a typical day
        radiation = np.random.normal(mean_radiation, stddev_radiation)

        # Calculate the energy generated by the panels
        energy = panel_area * panel_efficiency * radiation

        # Add the result to the list of results
        results.append(energy)

# Calculate the average energy generated over the year
avg_energy = np.mean(results) * days_per_year

# Print the results
print("Average energy generated over the year: {:.2f} kWh".format(avg_energy))
print("Average energy generated over one month: {:.2f} kWh".format(avg_energy/12))
```

```python
plt.hist(results, bins=12)
plt.xlabel('Investment Value after 1 Year')
plt.ylabel('Frequency')
plt.show()
```
